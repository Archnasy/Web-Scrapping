```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

# College Wbsite
url = 'https://www.thrissurkerala.com/list-all-kerala-courses.aspx'

# Send an HTTP request to the URL
response = requests.get(url)

# Check if the request was successful (status code 200)
if response.status_code == 200:
    # Parse the HTML content of the page using BeautifulSoup
    soup = BeautifulSoup(response.text, 'html.parser')

    lst = []
    for link in soup.find_all('a'):
        lst.append(link.get('href'))
        #print(link.get('href'))

    url_lst = lst[1:41]

base_url = "https://www.thrissurkerala.com/"

name_lst = []
address_lst = []
phone_lst = []
website_lst = []

for i in url_lst:
    url = base_url + i

    # Send an HTTP request to the URL
    response = requests.get(url)

    # Check if the request was successful (status code 200)
    if response.status_code == 200:
        # Parse the HTML content of the page using BeautifulSoup
        soup = BeautifulSoup(response.text, 'html.parser')

        name_college = soup.find('span', {'id': 'ctl00_ContentPlaceHolder1_DetailsView1_Label2'})
        clg_address = soup.find('span', {'id': 'ctl00_ContentPlaceHolder1_DetailsView1_Label3'})
        phn = soup.find('span', {'id': 'ctl00_ContentPlaceHolder1_DetailsView1_Label4'})
        web_site = soup.find('a', {'id': 'ctl00_ContentPlaceHolder1_DetailsView1_HyperLink1'})

        name_lst.append(name_college.string)
        address_lst.append(clg_address.string)
        phone_lst.append(phn.string)
        website_lst.append(web_site.string)

    else:
        print(f"Failed to retrieve the web page. Status code: {response.status_code}")

new_dict = dict(College_Name=name_lst, Address=address_lst, Phone=phone_lst, WebSite=website_lst)

college_data = pd.DataFrame(new_dict)
new_data = college_data.set_index('College_Name')


# college_data.to_excel(file_name)
new_data.to_excel('College_Details.xlsx')

```
